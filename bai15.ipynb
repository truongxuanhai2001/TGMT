```python
from scipy import linalg
import numpy as np
```

```python
#Định nghĩa hpt tuyến tính
A = np.array([[2,1],[3,2]])
B = np.array([5,7])

#GIải hpt tuyến tính
x = linalg.solve(A, B)
print("Kết quả:",x)
```

```python
#Tính tích phân của 1 hàm số
from scipy import integrate

#Định nghĩa f(x)=x^2
def my_function(x):
    return x**2;

#Tính thích phân của f(x) từ 0 đến 3
integral = integrate.quad(my_function, 0, 3);
print("Tích phân f(x) từ 0 đến 1:", integral)
```

```python
#!pip install scipy
from scipy import linalg
import numpy as np

# Tính giá trị riêng và vector riêng của một ma trận:
A = np.array([[1,11], [20,1]])

# Tính giá trị riêng và vector riêng của ma trận
evals, evecs = linalg.eig(A)

print("Giá trị riêng: ", evals)
print("Vector riêng: ", evecs)
```
