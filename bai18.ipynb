```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
```

```python
image_path= 'E:/TGMT/anh/anhsen2.jpg'
image = Image.open('E:/TGMT/anh/anhsen2.jpg')
# Chuyển sang ảnh xám
gray_image = image.convert('L')

image_array = np.array(gray_image)

# Chuyển sang ảnh nhị phân
binary_image = 1*(image_array<200)

#thực hiện opening để loại bỏ nhiễu và nối các đối tượng
opened_image = ndimage.binary_closing(binary_image, structure=np.ones((7,7)))

# Đánh nhãn
labeled_image, num_features = ndimage.label(binary_image)

# Tạo lưới subplot với 4 dòng và 1 cột
fig, axs = plt.subplots(4, 1, figsize=(50, 50))

plt.subplot(4,1,1)
plt.imshow(gray_image, cmap='gray')
plt.title("Ảnh xám")

plt.subplot(4,1,2)
plt.imshow(binary_image, cmap='gray')
plt.title("Ảnh nhị phân")

plt.subplot(4,1,3)
plt.imshow(labeled_image, cmap='gray')
plt.title(f"Đã đánh nhãn ({num_features} đối tượng)")

plt.show();
```
