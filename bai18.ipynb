```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
```

```python
image_path= 'E:/TGMT/anh/anhsen2.jpg'
image = Image.open('E:/TGMT/anh/anhsen2.jpg')
# Chuyn sang nh xm
gray_image = image.convert('L')

image_array = np.array(gray_image)

# Chuyn sang nh nh phn
binary_image = 1*(image_array<200)

#thc hin opening  loi b nhiu v ni cc i tng
opened_image = ndimage.binary_closing(binary_image, structure=np.ones((7,7)))

# nh nhn
labeled_image, num_features = ndimage.label(binary_image)

# To li subplot vi 4 dng v 1 ct
fig, axs = plt.subplots(4, 1, figsize=(50, 50))

plt.subplot(4,1,1)
plt.imshow(gray_image, cmap='gray')
plt.title("nh xm")

plt.subplot(4,1,2)
plt.imshow(binary_image, cmap='gray')
plt.title("nh nh phn")

plt.subplot(4,1,3)
plt.imshow(labeled_image, cmap='gray')
plt.title(f" nh nhn ({num_features} i tng)")

plt.show();
```
